# **Algoritmo Quick-Union**

Utilizamos la libreria graphviz y la funcion drawUF(lista de ids) para visualizar que nodos estan conectados (los que tienen el mismo id).

```python
import graphviz as gv

def drawDS(ds):
  graph = gv.Digraph("DisjointSet")
  graph.graph_attr['rankdir'] = "BT"
  for e, p in enumerate(ds):
    graph.node(str(e))
    if p != e:
      graph.edge(str(e), str(p))
  return graph
```

```python
#Definimos la funcion find
#devuelve la raiz o padre de a
def find(s, a):
    i = a
    while s[i] != i:
        i = s[i]
    return i
```

```python
#Definimos la funcion union
def union(s, a, b):
    #Raiz o padre de a
    pa = find(s, a)
    #print(pa)
    #Raiz o padre de b
    pb = find(s, b)
    #print(pb)
    print ("Despues del find estos son los nuevos valores de id[p] =", pa, " id[q]=", pb)
    #Solo actualizamos un valor que es el ultimo en cuanto a la raiz del padre (ya no se hace el for ni se cambia todos los ids de q para los de p como en Quick-find)
    s[pa] = pb
    #print(s)
```

```python
#Visializamos una lista de elementos de 0 a 9
s = [i for i in range(10)]
print(s)

#ds = [0, 1, 2, 2, 8, 5, 5, 7, 8, 2]
drawDS(s)
```

```python
# Unimos 3 y 4, la raiz de 3 es ahora 4
union(s, 3, 4)
print(s)
drawDS(s)
```
