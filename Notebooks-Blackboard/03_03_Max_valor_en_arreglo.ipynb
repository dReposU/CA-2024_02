```python
def obtenerMax(arr, i, j):
  if i == j:
    return arr[i]

  med = (i+j) // 2
  maxi = obtenerMax(arr, i, med)
  maxd = obtenerMax(arr, med + 1, j)
  return maxi if maxi > maxd else maxd
```

```python
arr = [random.randint(1,200) for _ in range(25)]
print(arr)
```

```python
print(max(arr))
```

```python
max_valor = obtenerMax(arr, 0, len(arr)-1)
print(max_valor)
```

```python
assert max_valor == max(arr)
print(max_valor)
```
