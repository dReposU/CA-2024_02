Escriba un algoritmo en Python que implemente una funcion recursiva llamada contador_palabras que reciba tres parametros: un arreglo de palabras, la posicion inicial, y final del arreglo. La funcion deberá retornar el numero de palabras del arreglo.

Ejemplo:

texto="Lorem ipsum dolor sit amet, consectetur adipiscing elit. Nam arcu arcu, euismod mollis rhoncus quis, egestas at leo. Maecenas luctus neque nulla, vel imperdiet lacus ultrices vel. Mauris eget eros vel lacus molestie blandit."

Input
-------
contador_palabras(arreglo,0,longitud_del_arreglo)

Output
---------
34

# Utilizando split para obtener el arreglo de palabras

```python
def contador_palabras(arreglo, inicio, final):
    # Caso base: si inicio es mayor que final, no hay palabras que contar
    if inicio > final:
        return 0
    # Caso recursivo: cuenta la palabra actual y avanza al siguiente
    return 1 + contador_palabras(arreglo, inicio + 1, final)

# Ejemplo de uso
texto = "Lorem ipsum dolor sit amet, consectetur adipiscing elit. Nam arcu arcu, euismod mollis rhoncus quis, egestas at leo. Maecenas luctus neque nulla, vel imperdiet lacus ultrices vel. Mauris eget eros vel lacus molestie blandit."
arreglo = texto.split()  # Dividimos el texto en palabras
longitud_del_arreglo = len(arreglo) - 1  # Índice de la última palabra

# Llamada a la función
numero_de_palabras = contador_palabras(arreglo, 0, longitud_del_arreglo)
print(numero_de_palabras)
numero_de_palabras = contador_palabras(arreglo, 0, 2)
print(numero_de_palabras)
```

# Partiendo de un string (sin split)

```python
def contador_palabras(texto, inicio, final):
    # Caso base: si el inicio es mayor que el final, no hay palabras que contar
    if inicio > final:
        return 0

    # Identificar el final de una palabra
    while inicio <= final and texto[inicio] == ' ':
        inicio += 1

    if inicio > final:
        return 0

    while inicio <= final and texto[inicio] != ' ':
        inicio += 1

    # Llamada recursiva para contar la siguiente palabra
    return 1 + contador_palabras(texto, inicio, final)

# Ejemplo de uso
texto = "Lorem ipsum dolor sit amet, consectetur adipiscing elit. Nam arcu arcu, euismod mollis rhoncus quis, egestas at leo. Maecenas luctus neque nulla, vel imperdiet lacus ultrices vel. Mauris eget eros vel lacus molestie blandit."
longitud_del_texto = len(texto) - 1
print(f"{longitud_del_texto}")

# Llamada a la función
numero_de_palabras = contador_palabras(texto, 0, longitud_del_texto)
print(numero_de_palabras)
numero_de_palabras = contador_palabras(texto, 0, 16)
print(numero_de_palabras)
```
