# Manual solution (educational) drawings of graphviz

```python
import graphviz as gv
```

```python
def draw_graph(dot_code):
    graph = gv.Source(dot_code)
    return graph
```

```python
dot_code = """
graph G {
    rankdir=LR
    bgcolor = black
    node[color=white fontcolor=orange width=.6]
    edge[color= "#a1a1a1" fontcolor = "#a1a1a1" ]
    
    A [xlabel="∞"];
    B [xlabel="∞"];
    C [xlabel="∞"];
    D [xlabel="∞"];
    E [xlabel="∞"];
    F [xlabel="∞"];
    G [xlabel="∞"];
    H [xlabel="∞"];

    
    A -- B [label="31"];
    A -- C [label="36"];
    A -- G [label="43"];
    
    B -- C [label="37"];
    B -- D [label="35"];
    B -- E [label="38"];
    B -- F [label="44"];
    B -- G [label="40"];
    
    C -- D [label="33"]; 
    C -- E [label="53"];
    C -- H [label="58"];
    
    D -- E [label="42"];
    
    E -- F [label="48"];
    E -- H [label="41"];
    
    F -- G [label="39"];
    F -- H [label="46"];
    
    G -- H [label="44"];
}
"""
draw_graph(dot_code)
```
