```python
from graphviz import Graph
lag={
'A': ['B', 'C'],
'B': ['A', 'D', 'E'],
'C': ['A', 'F'],
'D': ['B'],
'E': ['B', 'F'],
'F': ['C', 'E']
}
# 1. crear un grafo
grafo=Graph()
#1. agragar los nodos al grafo
for nodo in lag.keys():
    grafo.node(nodo)
#2. establecemos las las aristas al grafo
for nodo in lag.keys():
    for vecino in lag[nodo]:
        if nodo<vecino:
            grafo.edge(nodo, vecino)
#pintamos
grafo
```
